```python
import json
from collections import Counter
import random
```

```python
def getAnswer(name):
    path = name + ".json"
    with open(path, 'r') as f:
        data = json.load(f)

    # Преобразование данных из словаря в список

    arr = [round(num, 6) for num in data if isinstance(num, (int, float))]
    # Нахождение самых часто встречающихся чисел
    counts = Counter(arr)
    max_count = max(counts.values())

    # Определение индекса первого появления числа и его количество повторений
    result = []
    for num in counts:
        index = arr.index(num)
        count = counts[num]
        if count == max_count:
            result.append([index, num, count])

    print(result)
    path = name + "_Answer.json"
    with open(path, 'w') as f:
        json.dump(result, f)
```

```python
def randomFillJson(name, n):
    random_numbers = [random.uniform(-1000000, 1000000) for _ in range(n)]
    a = random.uniform(-1000000, 1000000)
    random_numbers.append(a)
    random_numbers.append(a)
    #random_numbers.append("asdsad")
    # Запись массива в файл JSON без ключа
    output = [num for num in random_numbers]
    path = name + ".json"
    with open(path, 'w') as f:
        json.dump(output, f)
```
